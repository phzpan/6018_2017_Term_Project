# SQLite

```python
def save_to_sqlite(sqlite_file, table_name):
    
    import pandas as pd
    import sqlite3

    #df = sat_df()
    df = pd.read_csv("States.csv", delimiter=",")
    print(df.head(5))

    conn = sqlite3.connect(sqlite_file)
    #cur = conn.cursor()

    df.to_sql(table_name, conn, if_exists="replace")
    conn.close()
    return None

def read_from_sqlite(sqlite_file, table_name):
    
    import pandas as pd
    import sqlite3
    try:    
        conn = sqlite3.connect(sqlite_file)
        #cur = conn.cursor()    

        query = """select * from """ + table_name
        df_sql = pd.read_sql(query, conn)
        df_sql = df_sql.drop('index', 1)
        #conn.commit()
    except:    
        pass
    finally:    
        if conn:
            conn.close()
    
    return df_sql
    
sqlite_file = 'states_sat.db'    
table_name = 'sat_table'   
save_to_sqlite(sqlite_file, table_name)
df=read_from_sqlite(sqlite_file, table_name)
print('- '*30)
print(df.head(5))

```
